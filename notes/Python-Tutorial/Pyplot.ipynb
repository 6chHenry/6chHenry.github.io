```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
matplotlib.rcParams['font.sans-serif'] = ['KaiTi']  # 设置中文字体
x = np.arange(5)
y = np.random.rand(5)

fig, ax = plt.subplots(1, 3, figsize=(9,3))

ax[0].plot(x, y, 'o')
ax[0].set_title("fmt='o'")

ax[1].plot(x, y, marker='o')          # 直线 + marker
ax[1].set_title("marker='o'")

ax[2].plot(x, y, marker='o', linestyle='None')
ax[2].set_title("标记='o' + linestyle='None'")

plt.tight_layout()
plt.show()
```
